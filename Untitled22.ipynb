```python
#https://economictimes.indiatimes.com/archivelist/year-2016,month-9,starttime-42615.cms
archive_url = []
i=1
k = 42614
while i <31:
    
    archive_url.append('https://economictimes.indiatimes.com/archivelist/year-2016,month-9,starttime-'+str(k)+'.cms')
    i = i+1
    k=k+1
archive_url
```

```python
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

for url in archive_url[1:5]:
    driver.get(url)
    news_link = driver.find_elements_by_xpath("//td[@class='contentbox5']//ul//li//a")
    urls = []
    for element in news_link:
        urls.append(element.get_attribute('href'))
    date = []
    author=[]
    vertical = []
    headline = []
    desc = []
    for i in urls:
        driver.get(i)

        current_url = driver.current_url
        
        try:
            date.append(driver.find_element_by_xpath("//time").text.replace('Updated: ','').replace('Created: ','').replace('Last ',''))
        except:
            date.append('-')
        try:
            
            desc.append(driver.find_element_by_xpath("//div[@class='artText medium'][1]").text.replace('\n',' '))
        except:
            desc.append('-')
        try:
            if driver.find_element_by_xpath("//div[@data-brcount='0']//p[1]").text[:2] == 'By':
                author.append(driver.find_element_by_xpath("//div[@data-brcount='0']//p[1]").text)
            else:
                author.append('ET Bureau')
        except:
            try:
                author.append(driver.find_element_by_xpath("//div[@class='dtc vam artByline']").text)
            except:
                author.append('-')
            
        try:
            headline.append(driver.find_element_by_xpath("//h1[@class='artTitle font_faus']").text)
        except:
            headline.append('-')
            
        if len(current_url.split('/')) == 8:
            vertical.append(current_url.split('/')[4])
        else:
            vertical.append(current_url.split('/')[3])

    df = pd.DataFrame()
    df_name = url.split('-')[3].replace('.cms','')
    df['Date'] = date
    df['Author'] = author
    df['Vertical'] = vertical
    df['Headline'] = headline
    df['Description'] = desc
   
```

```python
combined_csv = combined_csv[combined_csv['Date'] != '-']
combined_csv.reset_index(inplace=True,drop=True)
```
